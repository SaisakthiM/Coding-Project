Analysing a Numpy Array

```python
c = np.random.randint(10,100,(5,4))
print(c)
```

```python
# To find an Array Dimension

print(c.shape)
```

```python
# To Find Number Of Dimension

print(c.ndim) # Or
print(len(c.shape))
```

```python
# To find number of elements in an array

print(c.size) # Or
counter = 1
a = list(c.shape)
for x in a:
  counter *= x
print(counter)
```

```python
# Checking the data types of the values in the array

print(c.dtype)
```

Array Manipulation

```python
array = np.random.randint(1,9,(5,6))
print(array)
print(array.shape)
```

```python
# Find Transpose

trans = np.transpose(array)
print(trans)
print(trans.shape) # Or

trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# Reshaping a array
# this only applies if a*b (old row and column) == c*d (new row and column)
# in thus example , we have array which total elemebts is 30 and it has (5,6) row and column.
# so we can only reshape array which are multiole of 30 . Eg , 5-6 == 3*10
f = array[0][0]
print(f)

j = array.reshape(15,2)
print(j)
```
