```python
# Analyse the distribution of data in salary

fig, axis = plt.subplots(figsize=(8,8))
sns.distplot(i.salary)
```

The Above Graph is an example of "Skew" graph.
In this above graph , we cannot use mode due to imbalanced or inequal distribution of datas which actually incerases mean but fails to serve the purpose the equality of data

Repalcing Missing values with median

```python
# for filling , we are going to use fillna which automatically fills the NaN values with given values

i['salary'].fillna(i['salary'].median(),inplace=True)
```
