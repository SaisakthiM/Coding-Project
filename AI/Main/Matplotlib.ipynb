Matplotlib :

It is a module in python used for Plotting Graphs

```python
# Importing matplotlib

import matplotlib.pyplot as plt
```

```python
# Importing Datas For Working

import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes

diabeties_df = load_diabetes
x = np.linspace(0,10,100)
print(x)
```

```python
# Function To Find Sin Of All Elements Inside An Numpy Array

y = np.sin(x)
print(y)
z = np.cos(x)
print(z)
```

##Plotting The Data

```python
# Sin Wave

plt.plot(x,y)
plt.show()
```

```python
# Cos Or Cosine Wave :

plt.plot(x,z)
plt.show()
```

```python
# Tan Wave

plt.plot(y,z)
plt.show()
```

```python
# Cosec Wave
# Sin Wave

plt.plot(y,x)
plt.show()
```

```python
# Sec Wave

plt.plot(z,x)
plt.show()
```

```python
# Cot Wave

plt.plot(z,y)

plt.show()
```

```python
# Adding Title For X and Y Axis

plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sin Wave')
```

Drawing a Parabola

```python
# Parabola is Bowl shape graph and formula for it is ax +bx^2
x = np.linspace(-10,10,20)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(10,-10,20)

y = -(x**2)
plt.plot(x,y)
plt.show
```

```python
# Ther is feature Where we Can Not Only Draaw or plot graph through line bit with any character and colour through '(colour First Character(character)' after x and y
# here i used * symbol to plot points
# Default Character of plot is 'r-'
plt.plot(x,y, 'bx')
plt.show()
```

```python
n = np.linspace(-2,20,2)
plt.plot(n,np.sin(n), 'g-')
plt.plot(n,np.cos(n), 'r--')
plt.show()
```

Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lng = ['English', 'French', 'Spanish', 'Latin', 'Gerean']
people = [130, 100, 134, 120, 199]
ax.bar(lng, people)
plt.xlabel("Languages")
plt.ylabel("People Speaking")
plt.show()
```

Pie Chart

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lng = ['English', 'French', 'Spanish', 'Latin', 'Gerean']
people = [130, 100, 134, 120, 199]
ax.pie(people, labels=lng, autopct='%1.1f%%')
plt.xlabel("Languages")
plt.ylabel("People Speaking")
plt.show()
```
