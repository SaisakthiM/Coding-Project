House Price Datasets

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
california_housing = fetch_california_housing()
```

```python
raw_df.head()
```

```python
raw_df.boxplot()
plt.show()
```

```python
house = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
house['Price'] = california_housing.target
house.head()
```

#Distribution Plot

```python
import seaborn as sns
sns.set_theme()
sns.distplot(house['Price'])
```

Correlation : Lets take 2 variable x and y
1) Postive Correlation
If x ∝ y it is Postive Correlation
2) Negative Correlation
If x ∝ 1/y it is negative correlation

```python
corelation = house.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corelation,cbar=True, square=True, annot=True,annot_kws={'size': 8},cmap='Reds')
```
