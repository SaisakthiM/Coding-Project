# Django AI Notebook Example
This notebook sets up Django and accesses the `Document` model.

```python
import os
import sys
import django

# Go up one level from 'notebook/' and into 'src/'
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
src_path = os.path.join(project_root, 'src')

if src_path not in sys.path:
    sys.path.insert(0, src_path)

# Set up Django environment
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'django_ai.settings')
django.setup()

print("Django setup completed.")
```

```python
# Now import your model
from django_ai.documents.models import Document

# Access data
documents = Document.objects.all()
print(documents)
```
